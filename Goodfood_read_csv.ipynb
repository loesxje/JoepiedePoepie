```python
import pandas as pd
import numpy as np
import csv
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('recipes_v3.csv', sep=';')
data
```

```python
headers = list(data)
data = data.fillna(0);
```

```python
prep_time = data['prep_time']
cook_time = data['cook_time']
carbs = data['carbs']
fat = data['fat']

cluster_data = data[["prep_time" , "cook_time", "carbs", "fat"]]
```

```python
for i in range(len(carbs)):
    if carbs[i] == max(carbs):
        print(data['title'][i])
```
