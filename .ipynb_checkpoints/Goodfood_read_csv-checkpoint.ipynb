### Uitschieters verwijderen

```python
def reject_outliers(data, alpha):
    mu = np.mean(data)
    std = np.std(data)
    filtered = [i for i in data if (mu - alpha*std < i < mu + alpha*std)]
    return filtered
```

```python
filtered = reject_outliers(carbs, 1)
```

```python
for i in range(len(filtered)):
    if i < 2000:
        print(data['title'][i], i, carbs[i])
```
