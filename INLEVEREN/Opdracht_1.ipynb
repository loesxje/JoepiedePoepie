### Uitschieters verwijderen

```python
def reject_outliers(data, alpha):
    mu = np.mean(data)
    std = np.std(data)
    filtered = [i for i in data if (mu - alpha*std < i < mu + alpha*std)]
    return filtered
```

```python
filtered = reject_outliers(data['carbs'], 1)
```

### Stopwoorden wegfilteren


```python
stop_words = list(stopwords.words('english'))
```

```python
def filter_stop_words(col_name, data):
    data_filtered = []
    for sentence in data[col_name]:
        col = []
        for w in sentence.split():
            if not w in stop_words:
                col.append(w)
        data_filtered.append(col)
    return data_filtered
```

```python
data_filtered_title = filter_stop_words('title', data)
data_filtered_descr = filter_stop_words('description', data)
data_filtered_ingr = filter_stop_words('ingredients', data)
data_filtered_meth = filter_stop_words('method',data)
```

```python
def conc_words(data):
    list = []
    for sentence in data:
        list.append(' '.join(word for word in sentence))
    return list
```

```python
title = conc_words(data_filtered_title)
descr = conc_words(data_filtered_descr)
ingr = conc_words(data_filtered_ingr)
meth = conc_words(data_filtered_meth)
```

```python
df = pd.DataFrame({'title':title, 'description':descr, 'ingredients':ingr, 'method':meth})
```

## Data pre-processen

### Bag of Words

```python
def create_bagofwords(data):
    cnt_vect = CountVectorizer()
    bow = cnt_vect.fit_transform(data)
    tfidf_transformer = TfidfTransformer()
    bow_tfidf = tfidf_transformer.fit_transform(bow)
    
    return bow_tfidf
```

```python
tfidf_title = create_bagofwords(df['title'])
tfidf_descr = create_bagofwords(df['description'])
tfidf_ingr = create_bagofwords(df['ingredients'])
tfidf_meth = create_bagofwords(df['method'])
```

### Categorical string to int

```python
s = pd.Series(data['skill_level'].unique())
f_enc = pd.factorize(data['skill_level'])
```

## Clusteren

```python
features_words = np.hstack([
    tfidf_title.toarray(),
    tfidf_descr.toarray(),
    tfidf_ingr.toarray(),
    tfidf_meth.toarray(),
    f_enc[0].reshape(-1, 1),
])

features_int = np.hstack([
    np.array(data['carbs']),
    np.array(data['fat']),
    np.array(data['fibre']),
    np.array(data['kcal']),
    np.array(data['protein']),
    np.array(data['salt']),
    np.array(data['saturates']),
    np.array(data['sugars'])
])
```

### Clusterdata interpreteren
De titels van de gerechten worden per cluster bekeken, zo hopen we eruit te halen op welke redenenen het model geclusterd heeft.

```python
lst_0 = []
lst_1 = []
lst_2 = []
lst_3 = []
index = data.index
for i in range(len(data)):
    
    ind = index[i]
    if labels[i] == 0:
        lst_0.append(data['title'][ind])
    elif labels[i] == 1:
        lst_1.append(data['title'][ind])
    elif labels[i] == 2:
        
        lst_2.append(data['title'][ind])
    else:
        lst_3.append(data['title'][ind])
```
